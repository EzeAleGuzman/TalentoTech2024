Operaciones entre variables (aritméticas,relacionales, de asignación y operaciones lógicas )

## Aritméticas

```python

var1 = 15
var2 = 20
var3 = 30

suma = var1 + var2
print(suma)
```

```python
resta = var3 - var2
print(resta)
```

```python
multi = var1 * var2
print(multi)
```

```python
# división (cociente)
div = var3 / var2
print(div)
# división (parte entera)
div2 = var3 // var2
print(div2)
```

```python
# Obtener resto
resto = var3 % var2 # (30 módulo 20 = 10)
print(resto)
```
